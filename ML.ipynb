```python
def json_columnPrint(columnName):
    print("ColumnNumber  -  JSON")
    for x,y in zip(df.index, df[columnName]):
        print("{}  -  {}".format(x,y))
    return 0
    
```

```python
json_columnPrint("genres")
```

```python
series_genre = df.genres.apply(json.loads).apply(pd.io.json.json_normalize)#.pipe(lambda x: pd.concat(x.values)).head()
```

```python
df_genre = series_genre.iloc[0]
```

```python
count = 0
df_genre_total = pd.DataFrame([])
for x in range(len(series_genre)):
    #series_genre.iloc[x]
    id_tag = int(series_genre.index[x])
    df_genre = series_genre.iloc[x]
    id_list = list()
    id_list.append(id_tag) 
    id_list = id_list * df_genre.shape[0]
    
    df_genre['id_movie'] = pd.Series(id_list, index=df_genre.index)
    #df_genre.set_index = id_movie
    df_genre_total =  df_genre_total.append(df_genre, sort = True)
    count += 1
    
print(count)
```

```python
df_genre_total = df_genre_total.astype({'id_movie': 'int32'})
df_genre_total = df_genre_total.astype({'id': 'int32'})
df_genre_total = df_genre_total.drop('id', axis = 1)
df_genre_total.set_index('id_movie').head()
```

```python
df_genre_onehot = pd.get_dummies(df_genre_total.drop_duplicates().set_index('id_movie')).sum(level=0)
df_genre_onehot.columns = df_genre_onehot.columns.str.replace(r'name_', 'genre_')
```
